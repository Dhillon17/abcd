```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
X= np.array([[1,2],[1.5,1.8],[5,8],[8,8],[1,0.6],[9,11]])
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
def fit(data,k=2,max_iter=100):
    means=[]
    for i in range(k):
        means.append([data[i]])
    for i in range(max_iter):
        cluster=[]
        for j in range(k):
            cluster.append([])
        for point in data:
            distances=[((point-m)**2).sum() for m in means]
            minDistance = min(distances)
            l= distances.index(minDistance)
            cluster[l].append(point)
        change= False
        for j in range (k):
            new_mean= np.average(cluster[j],axis=0)
            if not np.array_equal(means[j], new_mean):
                change=True
            means[j]=new_mean
        if not change:
            break
    return means
```

```python
means=fit(X)
```

```python
def predict(test_data,means):
    predictions=[]
    for point in test_data:
        distances=[((point-m)**2).sum() for m in means]
        minDistance= min(distances)
        l= distances.index(minDistance)
        predictions.append(l)
    return predictions
```

```python
predict(X,means)
```
